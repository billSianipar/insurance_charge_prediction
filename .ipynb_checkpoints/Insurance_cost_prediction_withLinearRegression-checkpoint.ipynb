# Applying linear regression method.

**One of machine learning methods that could be used to attempt insurance charge prediction of a person in this case is linear regression.**

```python
insurance.head()
```

```python
sex = pd.get_dummies(insurance['sex'], drop_first=True)
smoker = pd.get_dummies(insurance['smoker'], drop_first=True)
region = pd.get_dummies(insurance['region'], drop_first=True)
```

```python
region.head()
```

**Based on this website some there are 4 categories of body mass index (bmi):**
https://www.nhlbi.nih.gov/health/educational/lose_wt/BMI/bmi-m.htm

- Underweight  :           bmi < 18.5
- Normal weight:   18.5 =< bmi =< 24.9
- Overweight   :    24.9 < bmi =< 29.9
- Obesity      :           bmi >= 30

**Based on the information right above, function below is created to categorize each body mass index of the 'bmi' column as Underweight, Normal weight, Overweight, and Obesity with codes: 1, 2, 3, 4 respectively.**

```python
def bmi_cat(cols):
    if cols < 18.5:
        return int(1)
    elif cols >= 18.5 and cols <=24.9:
        return int(2)
    elif cols > 24.9 and cols <=29.9:
        return int(3)
    elif cols >= 30:
        return int(4)
```

```python
insurance['bmiCAT'] = insurance['bmi'].apply(bmi_cat)
```

```python
insurance['bmiCAT'].head()
```

```python
insurance.head()
```

```python
bmiCAT = pd.get_dummies(insurance['bmiCAT'], drop_first=True)
```

```python
bmiCAT.head()
```

```python
insurance.head(2)
```

```python
insurance.drop(['sex','bmi', 'smoker', 'region', 'bmiCAT'], axis = 1, inplace=True)
```

```python
insurance.head()
```

```python
insurance = pd.concat([insurance,sex, bmiCAT,smoker ,region ], axis=1)
insurance.head()
```

### Data split for train and test data.

```python
X = insurance[['age', 'children', 'male', 2.0, 3.0, 4.0, 'yes', 'northwest', 'southeast', 'southwest']]
y = insurance['charges']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

### Apply linear regression to get a model.

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, y_train)
```

### Evaluate the model.

```python
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

**Get predictions using the X_test values from the data split.**

```python
predictions = lm.predict(X_test)
```

**Use jointplot to visualize the linear model, and also on of ways to get R-squared value.**

```python
from scipy import stats
def r2(x,y):
    return stats.pearsonr(x,y)[0]**2
sns.jointplot(x=y_test, y=predictions, kind='reg',stat_func=r2, height=7)
plt.xlabel('y_test')
plt.ylabel('predictions')
```

**Get error values as metrics to evaluate the regression.**

```python
from sklearn import metrics
print('R-squared: ', metrics.explained_variance_score(y_test, predictions))
print()
print('MAE: ', metrics.mean_absolute_error(y_test, predictions))
print('MSE: ', metrics.mean_squared_error(y_test, predictions))
print('RMSE: ', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

**It seems that the model that produced from this linear regression method is not really good. That can be concluded from the R-squared value and moreover by the error values.**

**The R-squared value indicates that the prediction values don't really match the real test values (the target, or in this case the insurance price). If the R-squared value is around 0.80 or 0.90 then the model might be good for predicting the insurance price.**

**An insurance charge can be between 1,000 to 60,000, so RMSE (root mean squared error) value of 6,000 is not good.**

**Hence this model shouldn't be used to predict insurance charge. Perhaps another machine learning method might be good to predict the insurance charge from this dataset.**
